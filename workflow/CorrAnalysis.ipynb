```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
```

```python
from scipy.stats import kendalltau, pearsonr, spearmanr
def kendall_pval(x,y):
    return kendalltau(x,y)[1]
def pearsonr_pval(x,y):
    return pearsonr(x,y)[1]
def spearmanr_pval(x,y):
    return spearmanr(x,y)[1]
def sig_pv(x):
    if x <= 0.05:
        return 1
    return 0
```

```python
fermented = pd.read_csv('../data/Fermented.csv').replace('-', 0 )
cheesy = pd.read_csv('../data/Cheesy.csv').replace('-', 0 )
milky = pd.read_csv('../data/Milky.csv').replace('-', 0 )
fruity = pd.read_csv('../data/Fruity.csv').replace('-', 0 )
df_merged = pd.read_csv('../data/Merged.csv')
```

```python
group_dict = {'Aldehyde':['Acetaldehyde','Benzaldehyde'],
              'Acid':['Acetic acid','Butanoic acid','Pentanoic acid','Hexanoic acid','Heptanoic acid','Octanoic acid',
                     'Nonanoic acid','Decanoic acid'],
              'Ketone':['2,3-Butanedione', '2,3-Pentanedione','2-Heptanone','Acetoin','2-Nonanone','2-Undecanone'],
             'Alcohol':['1-Hexanol','Phenol'],'Sulfur compound':['Dimethyl trisulfide','Dimethyl sulfone'],
             'Ester':['Ethyl acetate','delta-Decalactone']}
def get_group(cp, group_dict):
    for k in group_dict:
        if cp in group_dict[k]:
            return k
```

```python
def get_grouped(table, group_dict):
    samples = list(table.columns)[1:]
    gpd = pd.DataFrame( zip(samples), columns=['Samples'])
    for k in group_dict:
        temp_array = np.zeros( (len(samples)) )
        cp_list = group_dict[k]
        for cp in cp_list:
            if cp in list(table['Sample']):
                lst = list( table.iloc[ list(table['Sample']).index(cp) ] )[1:] 
                tp_list = [float(i) for i in lst]
                temp_array += np.array( tp_list )
        gpd[k] = temp_array
    return gpd
```

```python
fig, axs = plt.subplots(2,2,figsize=(16,20), sharey=True)
table_list = [fermented, cheesy, milky, fruity ]
titles = ['Fermented','Cheesy','Milky','Fruity']
for i in range(4):
    gpd = get_grouped( table_list[i] ,group_dict)
    classes, conc = [],[]
    for k in list( gpd.columns )[1:]:
        classes += [k]*len(gpd.index)
        conc += list(gpd[k])
    plot_table = pd.DataFrame(zip(conc,classes), columns=['Concentration','Class'])
    sns.boxplot(data=plot_table, x='Class', y='Concentration', ax= axs[int(i/2)][int(i%2)])
    sns.stripplot(data=plot_table, x='Class', y='Concentration',
                  edgecolor='black',linewidth=1, ax= axs[int(i/2)][int(i%2)])
    axs[int(i/2)][int(i%2)].set_yscale('log')
    axs[int(i/2)][int(i%2)].set_ylabel('Concentration',fontsize=12)
    axs[int(i/2)][int(i%2)].set_xlabel('')
    axs[int(i/2)][int(i%2)].grid()
    axs[int(i/2)][int(i%2)].set_title(titles[i], fontsize=16)
plt.savefig('../Figures/group_distribution.svg',format='svg',dpi=150)   
```

```python
group_colors = {'Aldehyde':'tab:blue', 'Acid':'tab:orange', 'Ketone':'tab:green', 
                'Alcohol':'tab:red', 'Sulfur compound':'tab:purple', 'Ester':'tab:brown'}
```

# Get correlation matrices

```python
def get_corr(table, G_name, group_dict):
    samples = list(table.columns)[1:]
    compounds = [ x  for x in list(table['Sample']) if x!= 'Label' ]
    temp_pd = table.copy()
    temp_pd = temp_pd.set_index(['Sample']).drop(['Label'],axis=0)
    temp_pd = temp_pd.T
    temp_pd = temp_pd[compounds]
    temp_pd = temp_pd.reset_index().drop(['index'],axis=1)
    temp_pd = temp_pd.astype(float)
    corr_pd = temp_pd.corr(method='pearson')
    sig_pd = temp_pd.corr( method= pearsonr_pval )
    sig_pd = sig_pd.applymap(sig_pv)
    vs = []
    for cp in list(sig_pd.columns):
        if sum( list(sig_pd[cp]) ) > 0:
            vs.append( cp )
    G=nx.Graph(name = G_name)
    G.add_nodes_from([(v,{"group": get_group(v, group_dict),
                          "color": group_colors[get_group(v, group_dict)] }) for v in vs])
    edge_list = []
    for i in range(len(sig_pd.index)):
        for j in range(i):
            if sig_pd.iloc[i,j] == 1:
                v1, v2 = list(sig_pd.index)[i], list(sig_pd.index)[j]
                weight = corr_pd.iloc[i,j]
                if weight > 0:
                    edge_list.append( (v1,v2,{'weight':abs(weight)*4,'edgecolor':'salmon'}) )
                else:
                    edge_list.append( (v1,v2,{'weight':abs(weight)*4,'edgecolor':'tab:blue'}) )
    G.add_edges_from(edge_list)              
    
    return G  
```

```python
fermented_corr = get_corr(fermented, 'Fermented', group_dict)
cheesy_corr = get_corr(cheesy, 'Cheesy', group_dict)
milky_corr = get_corr(milky, 'Milky', group_dict)
fruity_corr = get_corr(fruity, 'Fruity', group_dict)
```

```python
Gs =[ fermented_corr, cheesy_corr, milky_corr, fruity_corr ] 
fig, axs = plt.subplots(2,2,figsize=(26,26))
for i in range(4):
    G = Gs[i]
    nodecolors = [ group_colors[get_group(v, group_dict)] for v in G.nodes]
    edgecolors = [G[u][v]['edgecolor'] for u,v in G.edges]
    weights = [G[u][v]['weight'] for u,v in G.edges]
    nx.draw_circular(G, with_labels = True,edge_color=edgecolors,node_color=nodecolors, edgecolors='black',
                     font_weight='bold', width=weights,ax=axs[int(i/2)][int(i%2)])
    axs[int(i/2)][int(i%2)].set_title( G.name, fontsize=20)
plt.subplots_adjust(hspace=0.1,wspace=-0.001)
# plt.savefig('../Figures/sig_corr.svg',format='svg',dpi=150)
```
